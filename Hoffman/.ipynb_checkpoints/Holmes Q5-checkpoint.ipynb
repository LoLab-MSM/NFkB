```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x10y10 = np.loadtxt(open("holmes1010.csv", "rb"), delimiter=",")
x10y1 = np.loadtxt(open("holmes101.csv", "rb"), delimiter=",")
x1y10 = np.loadtxt(open("holmes110.csv", "rb"), delimiter=",")
x1y1 = np.loadtxt(open("holmes11.csv", "rb"), delimiter=",")
```

```python
plt.figure()

plt.plot(x10y10[0,:], x10y10[1,:])
plt.plot(x10y1[0,:], x10y1[1,:])
plt.plot(x1y10[0,:], x1y10[1,:])
plt.plot(x1y1[0,:], x1y1[1,:])
plt.xlabel("x", fontsize=16)
plt.ylabel("y", fontsize=16)
plt.ylim(ymin = 0, ymax =11)
plt.xlim(xmin = 0, xmax = 11)
plt.title('Equilibrium Analysis X vs. Y')
# plt.plot(datax, datay)
```

```python
fig = plt.figure(figsize=(15,5))
# fig.subplots_adjust(wspace = 1, hspace = 0.1)
ax1 = fig.add_subplot(1,4,1)
ax2 = fig.add_subplot(1,4,2)
ax3 = fig.add_subplot(1,4,3)
ax4 = fig.add_subplot(1,4,4)

ax1.plot(x10y10[0,:], 'r-', label='x')
ax1.plot(x10y10[1,:], 'b-', label='y')
#ax1.plot(z, 'g-', label='prey')
ax1.set_title("X10 vs Y10")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
# ax1.grid()
ax1.legend(loc='best')

ax2.plot(x10y1[0,:], 'r-', label='x')
ax2.plot(x10y1[1,:], 'b-', label='y')
ax2.set_title("X10 vs Y1")
ax2.set_xlabel("x")
ax2.set_ylabel("y")  
# ax2.set_title("X10 vs Y10 (b = 1)")

ax3.plot(x1y10[0,:], 'r-', label='x')
ax3.plot(x1y10[1,:], 'b-', label='y')
#ax1.plot(z, 'g-', label='prey')
ax3.set_title("X1 vs Y10")
ax3.set_xlabel("x")
ax3.set_ylabel("y")
# ax3.grid()
ax3.legend(loc='best')

ax4.plot(x1y1[0,:], 'r-', label='x')
ax4.plot(x1y1[1,:], 'b-', label='y')
#ax1.plot(z, 'g-', label='prey')
ax4.set_title("X1 vs Y1")
ax4.set_xlabel("x")
ax4.set_ylabel("y")
# ax4.g?rid()
ax4.legend(loc='best')

plt.tight_layout()
```
